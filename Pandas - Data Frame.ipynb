# Pandas

Pandas is a powerful open-source data analysis and manipulation library for Python. It provides easy-to-use data structures and functions for efficiently working with structured data such as tabular, time series, and matrix data.

## Import Pandas

```python
import pandas as pd #Imports the Pandas library into your Python code with the alias pd
import numpy as np #Imports the NumPy  library into your Python code with the alias np
```

Pandas has two objects, namely series and data frames.

# Data Frame
Data Frame is a collection of series, with at least one series.

## Convert Data Series into Data Frame

```python
#Data Series

dict_price = {"Book" : 2500,
             "Eraser" : 500,
             "Ruler" : 1200,
             "Pen" : 1500,
             "Pencil" : 1000}

dict_stock = {"Book" : 1000,
             "Eraser" : 200,
             "Ruler" : 100,
             "Pen" : 300,
             "Pencil" : 500}
```

```python
#dictionary to series transformation

price = pd.Series(dict_price)
price
```

```python
#dictionary to series transformation

stock = pd.Series(dict_stock)
stock
```

```python
store = pd.DataFrame({'price': price, 'stock': stock})
store
```

```python
store['price']
```

```python
store['price']['Book':'Ruler'] #Explicit Index
```

```python
store['price'].iloc[0:4] #Implicit Index
```

### Get Specified Columns

```python
#Retrieve Specified Coloumn

df[['title','type','director','country','release_year']]
```

```python
#Another way to retrieve specific columns from the DataFrame

sel_columns = ['title','type','director','country','release_year']
df[sel_columns]
```

```python
#Filters DataFrame to select rows where the value in the 'release_year' column equals 1998.

df[df['release_year']==1998]
```

```python
#Filters DataFrame to select rows where the value in the 'release_year' column is greater than 2000.

df[df['release_year']>2000]
```

```python
#filters DataFrame rows where 'release_year' is greater than '2000' and 'type' column is 'Movie'.

df[(df['release_year']>2000) & (df['type']=='Movie')]
```

```python
#Filters DataFrame rows where the value in the 'release_year' column is greater than 1990 and less than 2010.

df[((df['release_year']>1990) & (df['release_year']<2010))]
```

```python
#Filters the rows where 'release_year' is greater than 1990 or the value in the 'type' column is 'Movie'.

df[['title','type','director','country','release_year']][(df['release_year']>1990) | (df['type']=='Movie')]
```
